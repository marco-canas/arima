**C√≥digo para Animaci√≥n en Jupyter**:  
```python
import matplotlib.animation as animation
from IPython.display import HTML

fig, ax = plt.subplots()
def animate(i):
    ax.clear()
    df.groupby('Grado').head(i).boxplot(column='Nota', by='Grado', ax=ax)
    return ax,

ani = animation.FuncAnimation(fig, animate, frames=5, interval=1000)
HTML(ani.to_jshtml())
```
