```python
import matplotlib.pyplot as plt
import numpy as np

# Semilla para reproducibilidad
np.random.seed(42)

# Crear datos artificiales (4 aos, 48 meses)
meses = np.arange(1, 49)
datos_reales = np.sin(meses / 6) * 50 + 100 + np.random.normal(0, 5, size=48)
predicciones = np.sin(meses / 6) * 50 + 100 + np.random.normal(0, 10, size=48)

# Graficar
plt.figure(figsize=(12, 6))
plt.plot(meses, predicciones, color='red', label='Predicciones', linewidth=2)
plt.plot(meses, datos_reales, color='blue', linestyle='--', label='Datos Reales', linewidth=2)

# Etiquetas y detalles
plt.title("Predicciones vs Datos Reales de Casos de Dengue (4 aos)", fontsize=14)
plt.xlabel("Meses", fontsize=12)
plt.ylabel("Nmero de casos", fontsize=12)
plt.legend()
plt.grid(alpha = 0.1)
plt.tight_layout()
plt.savefig(r"C:\Users\marco\Downloads\predicciones_vs_datos_reales.png", dpi=300)
plt.show()
```
