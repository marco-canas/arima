## 2.1 Datos epidemiológicos

* Casos por fecha
* Municipio
* Semana epidemiológica

Para la estimación del número de casos de dengue por semana epidemiológica, se utilizaron variables específicas provenientes de la base de datos suministrada por la Secretaría de Salud. En particular, se empleó la variable fecha de inicio de síntomas, la cual permitió ubicar temporalmente cada caso y realizar su posterior agregación semanal. Asimismo, se filtraron los registros según el municipio de procedencia, seleccionando únicamente los casos correspondientes a Caucasia, Antioquia.

Adicionalmente, se consideraron las variables de ubicación como barrio y vereda, con el fin de describir la distribución espacial de los casos dentro del municipio. Cabe resaltar que los datos originales se encontraban reportados como casos ocurridos en fechas específicas, por lo que fue necesario realizar una transformación temporal.

Este proceso se llevó a cabo mediante el uso de Python, donde las fechas de inicio de síntomas fueron convertidas a semanas epidemiológicas siguiendo los lineamientos establecidos por el Instituto Nacional de Salud (INS) de Colombia. Esto permitió estandarizar la información y obtener el número de casos agregados por semana, incluyendo aquellas semanas en las que no se presentaron casos, garantizando así la continuidad de la serie temporal.

En cuanto a las variables meteorológicas, estas fueron obtenidas a partir de la plataforma NASA POWER, de donde se extrajeron datos correspondientes a temperatura (promedio, máxima y mínima), precipitación, humedad (relativa y específica) y velocidad del viento (promedio, máxima y mínima). La selección de estas variables se fundamenta en su influencia sobre la dinámica del vector Aedes aegypti, particularmente en aspectos como su desarrollo, supervivencia y capacidad de dispersión.

A partir de la variable de precipitación, se derivó una variable adicional correspondiente al número de días con lluvia, definida como aquellos días con valores iguales o superiores a 1 mm de precipitación. Esta variable fue incluida debido a su relevancia en la formación de criaderos del vector, lo cual puede incidir en el incremento del riesgo de transmisión del dengue.
