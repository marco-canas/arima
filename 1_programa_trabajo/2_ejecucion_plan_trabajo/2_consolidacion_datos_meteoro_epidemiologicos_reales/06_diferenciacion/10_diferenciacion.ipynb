# Inicio del análisis  

En la identificación de un modelo de la forma

$$
SARIMAX(p,d,q)(P,D,Q)_s
$$

para la predicción de casos de Dengue, el **análisis de diferenciación** es la etapa donde se determina cuántas veces debe transformarse la serie para lograr **estacionariedad**, condición fundamental para los modelos de la familia ARIMA y Seasonal ARIMA.

La diferenciación se divide en **dos componentes**:

$$
d \quad \text{(diferenciación no estacional)}
$$

$$
D \quad \text{(diferenciación estacional)}
$$

---



# 1. Diferenciación no estacional $d$

La diferenciación no estacional elimina **tendencias de largo plazo** en la serie.

La transformación utilizada es:

$$
\nabla y_t = y_t - y_{t-1}
$$

o en términos generales

$$
\nabla^d y_t
$$

donde $d$ indica el número de diferenciaciones.

### Evidencia observada en tu serie

A partir de los gráficos exploratorios:

1. La serie original presenta **tendencia creciente marcada**.
2. La media de la serie cambia con el tiempo.
3. El test de estacionariedad (por ejemplo el Augmented Dickey-Fuller test) típicamente arroja:

$$
p > 0.05
$$

lo que indica **no estacionariedad**.

Después de aplicar una diferenciación:

$$
y'*t = y_t - y*{t-1}
$$

la serie:

* oscila alrededor de una media aproximadamente constante
* presenta menor tendencia.

Esto sugiere que:

$$
d = 1
$$

es suficiente para estabilizar la media.

En series epidemiológicas como las del dengue raramente se requiere:

$$
d > 1
$$

porque una segunda diferenciación suele introducir **ruido excesivo**.

---



# 2. Diferenciación estacional (D)

La diferenciación estacional elimina **patrones periódicos**.

La transformación se define como:

$$
\nabla_s y_t = y_t - y_{t-s}
$$

donde $s$ es el período estacional.

Para datos **semanales de dengue**:

$$
s = 52
$$

ya que el ciclo epidemiológico suele repetirse aproximadamente cada año.

---



### Evidencia de estacionalidad

En tu análisis previo se observaron:

1. Oscilaciones periódicas en la descomposición temporal.
2. Un componente estacional moderado en el gráfico **Seasonal**.
3. Posible correlación en rezagos cercanos al ciclo anual.

Esto sugiere que puede existir una estructura estacional anual.



Por lo tanto, es razonable evaluar:

$$
D = 1
$$

si se confirma la presencia de autocorrelaciones en:

$$
lag = 52
$$

en los gráficos ACF.

---



# 3. Forma general de las transformaciones

Cuando se aplican ambas diferenciaciones, la serie transformada es:

$$
\nabla^d \nabla_s^D y_t
$$

En tu caso probable:

$$
\nabla^1 \nabla_{52}^1 y_t
$$

# 7. Interpretación epidemiológica

La diferenciación aplicada implica que el modelo no predice directamente los **niveles absolutos de casos**, sino los **cambios en los casos de dengue** entre semanas.

Esto es coherente con la dinámica de transmisión del dengue en regiones tropicales como Caucasia en el departamento de Antioquia, donde la evolución de la enfermedad depende fuertemente de:

* condiciones climáticas recientes
* dinámica del vector
* acumulación de susceptibles.
