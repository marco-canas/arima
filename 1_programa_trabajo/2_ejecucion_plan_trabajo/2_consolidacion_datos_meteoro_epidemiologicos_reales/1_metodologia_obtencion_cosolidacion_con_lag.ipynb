**Observaciones:** 

La plataforma descarga cada variable de forma independiente, es decir, no permite seleccionar varias variables simultáneamente para generar un único archivo CSV. Por esta razón, fue necesario unificar todas las variables en una sola base de datos.

```python
Dame el script de Python con una función de python que lea los archivos csv de la carpeta en la ubicación "C:\Users\marco\Documentos\investigacion\Bases de datos\Datos meteorológicos\AMSC" y fusione estos csv en un solo dataframe df que muestre estos datos meteorológicos guardados en la carpeta en al dirección que te acabo de dar. Quítale al código anterior el que indique el archivo origen. Pero que el atributo `fecha`lo convierta a tipo datetime. Pero que este script a medida que lee cada archivo .csv, convierta el atributo `fecha` en tipo datetime para que el dataframe final se fisiones con pd.merge() con el criterio para fusionar que sea este atributo `fecha`. Por favor, ten en cuenta que el atributo sep de pd.read_csv() debe ser ';' y el encoding 'utf-8'. 
```

Aquí tienes un **script de Python completo y comentado**, con una función que **lee todos los archivos CSV** en la carpeta que indicas y los **fusiona en un solo DataFrame** llamado `df`.

El script usa `pandas` y `glob`, dos herramientas ideales para esta tarea.

---


```python
# %% Importar librerías
import pandas as pd
import glob
import os

# %% Definir función para leer y fusionar CSV por la columna 'fecha'
def fusionar_csv_por_fecha(ruta_carpeta):
    """
    Lee todos los archivos CSV en la carpeta especificada,
    convierte la columna 'fecha' a tipo datetime y fusiona
    los DataFrames por esa columna usando pd.merge().

    Parámetros:
        ruta_carpeta (str): Ruta de la carpeta con los archivos CSV.

    Retorna:
        pd.DataFrame: DataFrame combinado por la columna 'fecha'.
    """
    # Listar todos los archivos CSV
    archivos_csv = glob.glob(os.path.join(ruta_carpeta, "*.csv"))
    
    if not archivos_csv:
        print("⚠️ No se encontraron archivos CSV en la ruta especificada.")
        return pd.DataFrame()
    
    df_final = None

    for archivo in archivos_csv:
        try:
            df_temp = pd.read_csv(archivo, sep=';', encoding='utf-8')
            
            # Verificar si existe la columna 'fecha'
            if 'fecha' not in df_temp.columns:
                print(f"⚠️ El archivo {os.path.basename(archivo)} no tiene la columna 'fecha'. Se omite.")
                continue
            
            # Convertir 'fecha' a tipo datetime
            df_temp['fecha'] = pd.to_datetime(df_temp['fecha'], errors='coerce')
            
            # Eliminar filas con fechas no válidas
            df_temp = df_temp.dropna(subset=['fecha'])
            
            # Ordenar por fecha (opcional, pero útil)
            df_temp = df_temp.sort_values('fecha')
            
            # Fusionar usando 'fecha' como clave
            if df_final is None:
                df_final = df_temp
            else:
                df_final = pd.merge(df_final, df_temp, on='fecha', how='outer')
            
            print(f"✅ Archivo fusionado: {os.path.basename(archivo)} ({len(df_temp)} filas)")

        except Exception as e:
            print(f"❌ Error procesando {archivo}: {e}")
    
    # Ordenar el DataFrame final por fecha
    if df_final is not None:
        df_final = df_final.sort_values('fecha').reset_index(drop=True)
        print("✅ Fusión completada correctamente.")
    else:
        print("⚠️ No se generó ningún DataFrame válido.")

    return df_final


# %% Usar la función con la ruta específica
ruta = r"C:\Users\marco\Documentos\investigacion\Bases de datos\Datos meteorológicos\AMSC"
df = fusionar_csv_por_fecha(ruta)

# %% Mostrar información básica del DataFrame final
print(f"\nTotal de filas: {len(df)}")
print(f"Columnas: {list(df.columns)}")
```
