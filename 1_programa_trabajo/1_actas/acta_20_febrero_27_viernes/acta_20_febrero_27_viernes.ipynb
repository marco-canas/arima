# 3. Tema central de la sesión: Medidas de desempeño para modelos ARIMA

Se desarrolló estudio formal de:



## 3.1 Criterios de Información

### AIC — Akaike Information Criterion

$$
AIC = -2\log(L) + 2k
$$

Interpretación:

* Penaliza complejidad
* Permite comparar modelos no anidados
* Menor AIC → mejor balance ajuste–parsimonia

---



### BIC — Bayesian Information Criterion

$$
BIC = -2\log(L) + k\log(n)
$$

* Penalización más fuerte
* Favorece modelos más simples

---



### HQIC — Hannan-Quinn

$$
HQIC = -2\log(L) + 2k\log(\log(n))
$$

Compromiso entre AIC y BIC.

---



## 3.2 Métricas de error de predicción

Se explicó formalmente:



### MAE

$$
MAE = \frac{1}{n}\sum |y_t - \hat{y}_t|
$$



### RMSE

$$
RMSE = \sqrt{\frac{1}{n}\sum (y_t - \hat{y}_t)^2}
$$



### MAPE

$$
MAPE = \frac{100}{n}\sum \left|\frac{y_t - \hat{y}_t}{y_t}\right|
$$

Se discutió:

* Diferencia entre ajuste in-sample y capacidad predictiva
* Importancia de validación fuera de muestra
