## Objetivos espec√≠ficos de la sesi√≥n

1. Realizar pr√°ctica de codificaci√≥n en **Python** para graficar casos de dengue en los √∫ltimos cinco a√±os y compararlos con predicciones del √∫ltimo a√±o.


```python
import numpy as np # Generaci√≥n de datos
import pandas as pd # visualiza de manera tabular y Manejo de datos
import matplotlib.pyplot as plt # vusuakiza de manera grpafica 
from statsmodels.tsa.arima.model import ARIMA # nos da el algoritmo para construir el modelo predictivo. 
from statsmodels.tsa.statespace.sarimax import SARIMAX

# ---------------------------
# 1. Datos artificiales de casos de dengue (2020-2024, mensual)
# ---------------------------
np.random.seed(702) 
n = 60  # 5 a√±os * 12 meses
time = pd.date_range(start="2020-01", periods=n, freq="M")

# serie con tendencia + estacionalidad anual (12 meses) + ruido
casos_dengue = (
    20 + 0.8*np.arange(n) +               # tendencia creciente
    15*np.sin(2*np.pi*np.arange(n)/12) +  # estacionalidad anual
    np.random.normal(0, 5, n)             # ruido
)
casos_dengue = np.round(casos_dengue).astype(int)  # redondear a enteros

df = pd.DataFrame({"Fecha": time, "Casos": casos_dengue})
df.set_index("Fecha", inplace=True)

# ---------------------------
# 2. ARIMA
# ---------------------------
model_arima = ARIMA(df["Casos"], order=(2,1,2))
fit_arima = model_arima.fit()
pred_arima = fit_arima.predict(start=n-12, end=n-1, typ="levels")  # √∫ltimo a√±o

plt.figure(figsize=(10,5))
plt.plot(df.index, df["Casos"], label="Casos reales", color="blue")
plt.plot(pred_arima.index, pred_arima, label="Predicci√≥n ARIMA", color="red")
plt.title("Modelo ARIMA (2,1,2) - Casos de Dengue (2020-2024)")
plt.xlabel("Tiempo")
plt.ylabel("Casos de dengue")
plt.grid(True)
plt.legend()
plt.savefig('arima_dengue_cases.png', dpi=300)
plt.show()

# ---------------------------
# 3. SARIMA
# ---------------------------
model_sarima = SARIMAX(df["Casos"], order=(1,1,1), seasonal_order=(1,1,1,12))
fit_sarima = model_sarima.fit(disp=False)
pred_sarima = fit_sarima.predict(start=n-12, end=n-1)

plt.figure(figsize=(10,5))
plt.plot(df.index, df["Casos"], label="Casos reales", color="blue")
plt.plot(pred_sarima.index, pred_sarima, label="Predicci√≥n SARIMA", color="red")
plt.title("Modelo SARIMA (1,1,1)(1,1,1,12) - Casos de Dengue (2020-2024)")
plt.xlabel("Tiempo")
plt.ylabel("Casos de dengue")
plt.grid(True)
plt.legend()
plt.savefig('sarima_dengue_cases.png', dpi=300)
plt.show()

# ---------------------------
# 4. ARIMAX (con variable ex√≥gena artificial)
# ---------------------------
# Creamos variable ex√≥gena ficticia (ejemplo: precipitaci√≥n en mm)
exog = 100 + 30*np.sin(2*np.pi*np.arange(n)/6) + np.random.normal(0,10,n)

model_arimax = SARIMAX(df["Casos"], order=(1,1,1), exog=exog)
fit_arimax = model_arimax.fit(disp=False)
pred_arimax = fit_arimax.predict(start=n-12, end=n-1, exog=exog[n-12:])

plt.figure(figsize=(10,5))
plt.plot(df.index, df["Casos"], label="Casos reales", color="blue")
plt.plot(pred_arimax.index, pred_arimax, label="Predicci√≥n ARIMAX", color="red")
plt.title("Modelo ARIMAX (1,1,1) con precipitaci√≥n - Casos de Dengue (2020-2024)")
plt.xlabel("Tiempo")
plt.ylabel("Casos de dengue")
plt.grid(True)
plt.legend()
plt.savefig('arimax_dengue_cases.png', dpi=300)
plt.show()
```
