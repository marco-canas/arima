## 3. Desarrollo de la asesoría

Durante la sesión de asesoría se abordaron y desarrollaron las siguientes acciones:

### 3.1 Análisis de correlación de atributos meteorológicos rezagados

* Se dio inicio al proceso de **determinación de los atributos meteorológicos rezagados** que presentan mayor relación con la variable objetivo `num_casos`.
* Se discutió la importancia de este análisis como insumo para la **reducción de dimensionalidad** y la correcta especificación del modelo SARIMAX.

### 3.2 Definición de la notación para atributos rezagados

* Se acordó de manera formal la notación para nombrar los atributos meteorológicos con rezago temporal, definida como:

$$
\texttt{atributo_nasa_lag_i}, \quad i \in {1, 2, \ldots, 11, 12}
$$

* Se estableció que el índice $i$ representa la **i-ésima semana epidemiológica de rezago**, garantizando claridad semántica, trazabilidad del código y coherencia metodológica en los cuadernos Jupyter.

### 3.3 Métodos de correlación estadística

* Se acordó utilizar distintos enfoques de **correlación estadística**, con el fin de capturar diferentes tipos de relación entre variables:

  * **Correlación lineal de Pearson**, para evaluar relaciones lineales.
  * **Correlación de Kendall**, adecuada para relaciones monótonas y datos con posibles valores atípicos.
  * **Correlación de Spearman**, basada en rangos y útil ante posibles no linealidades.

* Se enfatizó la importancia de comparar estos coeficientes para obtener una visión más robusta del comportamiento de los datos.

### 3.4 Análisis gráfico mediante diagramas de dispersión

* Se definió la construcción de **diagramas de dispersión** para las parejas:

$$
(\texttt{num_casos}, ; \texttt{atributo_nasa_lag_i})
$$

* Se resaltó el valor didáctico y analítico de estas visualizaciones para identificar patrones, tendencias, posibles no linealidades y presencia de valores atípicos.
