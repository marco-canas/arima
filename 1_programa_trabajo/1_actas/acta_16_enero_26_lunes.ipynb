## 4. Etapas de consolidación del modelo SARIMAX

Se acuerdan las siguientes etapas:

### Etapa 1. Ajuste final del modelo SARIMAX

* Definición de los parámetros $(p, d, q)$ y $(P, D, Q, s)$.
* Inclusión definitiva de variables exógenas rezagadas.
* Análisis de significancia estadística de los coeficientes.



### Etapa 2. Diagnóstico del modelo

* Análisis de residuos (ruido blanco, normalidad, autocorrelación).
* Verificación de supuestos del modelo.

### Etapa 3. Validación y evaluación

* Separación de conjuntos de entrenamiento y prueba.
* Evaluación mediante métricas (RMSE, MAE, MAPE).
* Comparación con modelos base.

### Etapa 4. Predicción y análisis epidemiológico

* Generación de predicciones a corto y mediano plazo.
* Interpretación de resultados desde el contexto epidemiológico local.

### Etapa 5. Documentación y socialización

* Sistematización metodológica.
* Elaboración de gráficos y tablas finales.
* Preparación de resultados para informe o artículo académico.
