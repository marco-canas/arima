## 4. Evaluaciones que sí debes aplicar

### A. Diagnóstico de residuos

#### Prueba de Ljung-Box

Permite verificar si los residuos del modelo son independientes.

Interpretación:

* p > 0.05 → los residuos parecen ruido blanco → buen ajuste.
* p < 0.05 → aún existe autocorrelación → el modelo no capturó toda la estructura temporal.

Esto es obligatorio en modelos ARIMA/SARIMA.

---

### B. Criterios de información

#### AIC

Penaliza modelos muy complejos.

Menor AIC = mejor equilibrio entre ajuste y complejidad.

#### BIC

Similar al AIC, pero castiga más los modelos con demasiados parámetros.

Menor BIC = modelo más parsimonioso.

---

### C. Métricas predictivas

#### RMSE

Mide el error promedio cuadrático.

Más sensible a errores grandes.

#### MAE

Mide el error absoluto promedio.

Más robusto ante valores extremos.

#### MAPE

Mide error porcentual.

Muy útil para interpretar el error relativo.

Interpretación aproximada:

* <10% → excelente.
* 10–20% → bueno.
* 20–50% → aceptable.
* > 50% → deficiente.

```python
# ================================
# 1. LIBRERÍAS
# ================================
import pandas as pd
import numpy as np
import os

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.stats.diagnostic import acorr_ljungbox

import warnings
warnings.filterwarnings("ignore")
```

```python
# ================================
# 2. RUTAS
# ================================
ubicacion_datos_consolidados_janis = r"C:\Users\usuario1\OneDrive - Universidad de Antioquia\UNIVERSIDAD DE ANTIOQUIA\Proyecto SAT Dengue\Bases de datos\Modelos\modelo_pca_9_variables_rezagos_relevantes\Resultados"

ubicacion_imagenes_janis = r"C:\Users\usuario1\OneDrive - Universidad de Antioquia\UNIVERSIDAD DE ANTIOQUIA\Proyecto SAT Dengue\Bases de datos\Modelos\modelo_pca_9_variables_rezagos_relevantes\Gráficos"

os.makedirs(ubicacion_datos_consolidados_janis, exist_ok=True)
os.makedirs(ubicacion_imagenes_janis, exist_ok=True)
```

```python
# ================================
# 3. CARGA DE DATOS
# ================================
df = pd.read_excel(r"C:\Users\usuario1\OneDrive - Universidad de Antioquia\UNIVERSIDAD DE ANTIOQUIA\Proyecto SAT Dengue\Bases de datos\Modelos\modelo_pca_9_variables_rezagos_relevantes\datos_9_variables_rezagos_relevantes.xlsx")

df['fecha'] = pd.to_datetime(df['fecha'])
df.set_index('fecha', inplace=True)
```

```python
# ================================
# 4. VARIABLES
# ================================
y = df['casos_ln']

# Variables exógenas (PCA ya aplicado)
X = df.drop(columns=['casos_ln', 'año'], errors='ignore')
```

```python
# ================================
# 5. FUNCIONES DE MÉTRICAS
# ================================
def calcular_metricas(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    
    # evitar división por cero en MAPE
    y_true_safe = np.where(y_true == 0, 1e-6, y_true)
    mape = np.mean(np.abs((y_true - y_pred) / y_true_safe)) * 100
    
    return rmse, mae, mape


def ljung_box_test(residuos):
    pvalue = acorr_ljungbox(residuos, lags=[10], return_df=True)['lb_pvalue'].values[0]
    return pvalue

```

```python
# ================================
# 6. MODELOS
# ================================
modelos = [
    (1,1,0),
    (0,1,1),
    (1,1,1),
    (2,1,1),
    (1,1,2),
]

modelos_estacionales = [
    ((1,1,1),(1,1,1,52)),
    ((1,1,0),(1,1,1,52)),
    ((0,1,1),(1,1,1,52)),
]
```

```python
# ================================
# 7. PORCENTAJES TRAIN/TEST
# ================================
porcentajes = [0.8, 0.9, 0.95]

resultados_totales = []

```

```python
# ================================
# 8. LOOP PRINCIPAL
# ================================
for p in porcentajes:
    
    print("\n==============================")
    print(f"ENTRENAMIENTO {round(p*100)}% - TEST {round((1-p)*100)}%")
    print("==============================")
    
    split = int(len(df) * p)
    
    y_train, y_test = y[:split], y[split:]
    X_train, X_test = X[:split], X[split:]
    
    resultados = []
    errores = []

    # ---------- ARIMA ----------
    for order in modelos:
        try:
            model = SARIMAX(y_train, order=order)
            fit = model.fit(disp=False)
            
            pred = fit.forecast(steps=len(y_test))
            
            rmse, mae, mape = calcular_metricas(y_test, pred)
            ljung = ljung_box_test(fit.resid)
            
            resultados.append(["ARIMA"+str(order), fit.aic, fit.bic, rmse, mae, mape, ljung])
        
        except Exception as e:
            errores.append(f"ARIMA{order}: {e}")

    # ---------- SARIMA ----------
    for order, seasonal in modelos_estacionales:
        try:
            model = SARIMAX(y_train, order=order, seasonal_order=seasonal)
            fit = model.fit(disp=False)
            
            pred = fit.forecast(steps=len(y_test))
            
            rmse, mae, mape = calcular_metricas(y_test, pred)
            ljung = ljung_box_test(fit.resid)
            
            resultados.append(["SARIMA"+str(order)+str(seasonal), fit.aic, fit.bic, rmse, mae, mape, ljung])
        
        except Exception as e:
            errores.append(f"SARIMA{order}{seasonal}: {e}")

    # ---------- ARIMAX ----------
    for order in modelos:
        try:
            model = SARIMAX(y_train, exog=X_train, order=order)
            fit = model.fit(disp=False)
            
            pred = fit.forecast(steps=len(y_test), exog=X_test)
            
            rmse, mae, mape = calcular_metricas(y_test, pred)
            ljung = ljung_box_test(fit.resid)
            
            resultados.append(["ARIMAX"+str(order), fit.aic, fit.bic, rmse, mae, mape, ljung])
        
        except Exception as e:
            errores.append(f"ARIMAX{order}: {e}")

    # ---------- SARIMAX ----------
    for order, seasonal in modelos_estacionales:
        try:
            model = SARIMAX(y_train, exog=X_train, order=order, seasonal_order=seasonal)
            fit = model.fit(disp=False)
            
            pred = fit.forecast(steps=len(y_test), exog=X_test)
            
            rmse, mae, mape = calcular_metricas(y_test, pred)
            ljung = ljung_box_test(fit.resid)
            
            resultados.append(["SARIMAX"+str(order)+str(seasonal), fit.aic, fit.bic, rmse, mae, mape, ljung])
        
        except Exception as e:
            errores.append(f"SARIMAX{order}{seasonal}: {e}")
```

```python
# ================================
# 9. RESULTADOS
# ================================
df_res = pd.DataFrame(resultados, columns=[
        "Modelo", "AIC", "BIC", "RMSE", "MAE", "MAPE", "Ljung_Box_pvalue"
    ])
    
df_res = df_res.sort_values(by="AIC")

    # guardar
nombre = f"resultados_{int(p*100)}.xlsx"
df_res.to_excel(os.path.join(ubicacion_datos_consolidados_janis, nombre), index=False)

resultados_totales.append((p, df_res))

    # ================================
    # 10. CONSOLA
    # ================================
print("\nTOP 5 MODELOS:")
print(df_res.head(10))

print("\nModelos con error:")
for err in errores[:5]:  # solo muestra 5 para no saturar
    print(err)
```

```python
# ================================
# 11. RESUMEN FINAL
# ================================
print("\n==============================")
print("RESUMEN FINAL")
print("==============================")

for p, df_res in resultados_totales:
    print(f"\n--- {int(p*100)}% ---")
    print(df_res.head(3))
```
