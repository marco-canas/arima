# 11_identificacion_modelo

Proyecto ARIMA / ARIMAX
Modelación epidemiológica con variables meteorológicas.

# Inicio del análisis  

El **proceso de identificación de un modelo** de la familia

$$
SARIMAX(p,d,q)(P,D,Q)_s
$$

es una etapa fundamental dentro de la metodología de modelación de series temporales propuesta por George Box y Gwilym Jenkins en el enfoque conocido como Box-Jenkins methodology.



En el contexto de tu investigación para la predicción de casos de Dengue mediante un sistema de alertas tempranas, la **identificación del modelo** consiste en determinar los valores apropiados de:

$$
(p,d,q)(P,D,Q)_s
$$

a partir del análisis estadístico de la serie temporal.

A continuación se describe **todo el proceso metodológico de identificación**.

---



# 1. Verificación de estacionariedad

El primer paso consiste en determinar si la serie es **estacionaria**.

Una serie estacionaria cumple:

* media aproximadamente constante
* varianza aproximadamente constante
* autocorrelación dependiente solo del rezago.

### Herramientas utilizadas

1. inspección gráfica de la serie
2. media móvil y desviación estándar móvil
3. prueba de raíz unitaria como el
   Augmented Dickey-Fuller test.



### Resultado esperado

En tu serie de casos de dengue se observó:

* tendencia creciente
* cambios en la variabilidad.

Por lo tanto, la serie **no es estacionaria**.

---



# 2. Determinación del orden de diferenciación $d$

Para eliminar la tendencia se aplica diferenciación:

$$
y'_t = y_t - y_{t-1}
$$

Después de una diferenciación se obtiene una serie aproximadamente estacionaria.

Esto conduce a:

$$
d = 1
$$

---



# 3. Identificación de estacionalidad

Dado que los datos corresponden a **semanas epidemiológicas**, se evalúa la presencia de ciclos anuales.

La periodicidad esperada es:

$$
s = 52
$$

La evidencia de estacionalidad se obtiene mediante:

* descomposición temporal
* análisis del ACF en rezagos estacionales
* análisis visual de ciclos anuales.

Si existe dependencia en el rezago $52$, se considera diferenciación estacional:

$$
D = 1
$$

---



# 4. Análisis de la función de autocorrelación (ACF)

La **función de autocorrelación** mide la correlación entre la serie y sus rezagos.

Se calcula mediante herramientas de statsmodels en Python.

### Interpretación

* corte abrupto en ACF → componente **MA(q)**
* decaimiento gradual → componente **AR**

En tu gráfico se observó:

* un pico significativo en rezagos bajos
* luego las correlaciones se mantienen dentro del intervalo de confianza.

Esto sugiere un componente:

$$
q = 1
$$

---



# 5. Análisis de la función de autocorrelación parcial (PACF)

La **PACF** mide la correlación entre la serie y sus rezagos eliminando el efecto de rezagos intermedios.

Interpretación típica:

| Patrón                    | Modelo sugerido |
| ------------------------- | --------------- |
| Corte abrupto en PACF     | AR(p)           |
| Corte abrupto en ACF      | MA(q)           |
| Ambos decaen gradualmente | ARMA            |

En el análisis de tu serie se observa una estructura corta compatible con:

$$
p = 1
$$

---



# 6. Identificación de componentes estacionales

Si en el ACF aparecen picos en:

$$
lag = 52
$$

entonces existe dependencia estacional.

Esto permite identificar:

* componente autoregresivo estacional (P)
* componente de media móvil estacional (Q).

En muchos modelos epidemiológicos iniciales se prueba:

$$
P = 1
$$

$$
Q = 0
$$

---



# 7. Propuesta de modelos candidatos

Con base en el análisis previo se proponen varios modelos iniciales.

Ejemplos:

$$
SARIMAX(1,1,1)(1,1,0)_{52}
$$

$$
SARIMAX(0,1,1)(1,1,0)_{52}
$$

$$
SARIMAX(1,1,0)(1,1,0)_{52}
$$

$$
SARIMAX(1,1,2)(1,1,0)_{52}
$$

Estos modelos se estiman y se comparan usando criterios de información.

---



# 8. Selección del mejor modelo

Los modelos se comparan mediante:

* AIC (Akaike Information Criterion)
* BIC (Bayesian Information Criterion)

El mejor modelo es aquel que **minimiza estos criterios**.

---



# 9. Validación de residuos

Después de estimar el modelo se analiza si los residuos se comportan como **ruido blanco**.

Se verifica:

1. media cercana a cero
2. ausencia de autocorrelación
3. varianza constante.

Esto puede evaluarse con el **test de Ljung-Box**, disponible en statsmodels.

---



# 10. Resultado del proceso de identificación

Aplicando el proceso completo de identificación a tu serie semanal de casos de dengue, se obtiene como modelo candidato inicial:

$$
SARIMAX(1,1,1)(1,1,0)_{52}
$$

donde:

* (p=1) componente autoregresivo
* (d=1) diferenciación para eliminar tendencia
* (q=1) componente de media móvil
* (P=1) componente autoregresivo estacional
* (D=1) diferenciación estacional anual
* (Q=0)
* (s=52) semanas.

---



# 11. Interpretación epidemiológica

Este modelo sugiere que los casos de dengue en una semana dependen de:

1. la dinámica reciente de la enfermedad,
2. perturbaciones aleatorias recientes,
3. la dinámica estacional anual asociada al clima.

Esto es consistente con la transmisión del dengue en regiones tropicales como Antioquia y municipios como Caucasia.

---



# **Conclusión**

El proceso de identificación permite determinar una estructura preliminar del modelo SARIMAX que captura:

* tendencia temporal
* dependencia a corto plazo
* patrones estacionales anuales

de la serie de casos de dengue, proporcionando la base para la etapa posterior de **estimación y validación del modelo predictivo**.
