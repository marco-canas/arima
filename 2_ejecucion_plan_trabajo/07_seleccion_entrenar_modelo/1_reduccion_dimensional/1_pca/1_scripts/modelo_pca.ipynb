```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
import numpy as np

pca = PCA()
X_pca = pca.fit_transform(X_scaled)

varianza_acumulada = np.cumsum(pca.explained_variance_ratio_)
print(varianza_acumulada)
```

```python
n_componentes = np.where(varianza_acumulada >= 0.90)[0][0] + 1
print(f"Número de componentes: {n_componentes}")
```

```python
import pandas as pd

df_pca_final = pd.DataFrame(
    X_pca[:, :n_componentes],
    columns=[f'PC{i+1}' for i in range(n_componentes)]
)

df_pca_final['casos_ln'] = y.values
```

# Modelo PCA

```python
import numpy as np

# Variables
X = df_pca_final.drop(columns=['casos_ln'])
y = df_pca_final['casos_ln']
```

```python
def train_test_split_time_series(X, y, train_size):

    n = len(X)
    train_n = int(n * train_size)

    X_train = X.iloc[:train_n]
    X_test = X.iloc[train_n:]

    y_train = y.iloc[:train_n]
    y_test = y.iloc[train_n:]

    return X_train, X_test, y_train, y_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

def calcular_metricas(y_true, y_pred):

    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    mae = mean_absolute_error(y_true, y_pred)

    return rmse, mae, mape
```

```python
# ARIMA
arima_orders = [
    (1,1,0),
    (0,1,1),
    (1,1,1),
    (2,1,1),
    (1,1,2),
    (2,1,2)
]

# SARIMA
sarima_orders = [
    ((1,1,1),(1,1,1,52)),
    ((1,1,0),(1,1,1,52)),
    ((0,1,1),(1,1,1,52)),
    ((2,1,1),(1,1,1,52)),
    ((1,1,2),(1,1,1,52)),
    ((1,1,1),(0,1,1,52)),
    ((1,1,1),(1,1,0,52))
]

# ARIMAX (igual que ARIMA)
arimax_orders = [
    (1,1,0),
    (0,1,1),
    (1,1,1),
    (2,1,1),
    (1,1,2)
]

# SARIMAX
sarimax_orders = [
    ((1,1,1),(1,1,1,52)),
    ((1,1,0),(1,1,1,52)),
    ((0,1,1),(1,1,1,52)),
    ((2,1,1),(1,1,1,52)),
    ((1,1,2),(1,1,1,52))
]
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.stats.diagnostic import acorr_ljungbox
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

def evaluar_modelos(X_train, X_test, y_train, y_test):

    resultados = []

    # ======================
    # ARIMA
    # ======================
    for order in arima_orders:
        try:
            model = ARIMA(y_train, order=order)
            fit = model.fit()

            pred = fit.forecast(steps=len(y_test))

            rmse = np.sqrt(mean_squared_error(y_test, pred))
            mae = mean_absolute_error(y_test, pred)
            mape = np.mean(np.abs((y_test - pred) / y_test)) * 100

            lb = acorr_ljungbox(fit.resid, lags=[10], return_df=True)

            resultados.append({
                'Modelo': 'ARIMA',
                'Orden': order,
                'RMSE': rmse,
                'MAE': mae,
                'MAPE': mape,
                'AIC': fit.aic,
                'BIC': fit.bic,
                'Ljung_Box_pvalue': lb['lb_pvalue'].values[0]
            })
        except:
            pass

    # ======================
    # SARIMA
    # ======================
    for order, seasonal in sarima_orders:
        try:
            model = SARIMAX(y_train, order=order, seasonal_order=seasonal)
            fit = model.fit(disp=False)

            pred = fit.forecast(steps=len(y_test))

            rmse = np.sqrt(mean_squared_error(y_test, pred))
            mae = mean_absolute_error(y_test, pred)
            mape = np.mean(np.abs((y_test - pred) / y_test)) * 100

            lb = acorr_ljungbox(fit.resid, lags=[10], return_df=True)

            resultados.append({
                'Modelo': 'SARIMA',
                'Orden': order,
                'RMSE': rmse,
                'MAE': mae,
                'MAPE': mape,
                'AIC': fit.aic,
                'BIC': fit.bic,
                'Ljung_Box_pvalue': lb['lb_pvalue'].values[0]
            })
        except:
            pass

    # ======================
    # ARIMAX (con PCA)
    # ======================
    for order in arimax_orders:
        try:
            model = SARIMAX(y_train, exog=X_train, order=order)
            fit = model.fit(disp=False)

            pred = fit.forecast(steps=len(y_test), exog=X_test)

            rmse = np.sqrt(mean_squared_error(y_test, pred))
            mae = mean_absolute_error(y_test, pred)
            mape = np.mean(np.abs((y_test - pred) / y_test)) * 100

            lb = acorr_ljungbox(fit.resid, lags=[10], return_df=True)

            resultados.append({
                'Modelo': 'ARIMAX',
                'Orden': order,
                'RMSE': rmse,
                'MAE': mae,
                'MAPE': mape,
                'AIC': fit.aic,
                'BIC': fit.bic,
                'Ljung_Box_pvalue': lb['lb_pvalue'].values[0]
            })
        except:
            pass

    # ======================
    # SARIMAX
    # ======================
    for order, seasonal in sarimax_orders:
        try:
            model = SARIMAX(y_train, exog=X_train,
                            order=order, seasonal_order=seasonal)
            fit = model.fit(disp=False)

            pred = fit.forecast(steps=len(y_test), exog=X_test)

            rmse = np.sqrt(mean_squared_error(y_test, pred))
            mae = mean_absolute_error(y_test, pred)
            mape = np.mean(np.abs((y_test - pred) / y_test)) * 100

            lb = acorr_ljungbox(fit.resid, lags=[10], return_df=True)

            resultados.append({
                'Modelo': 'SARIMAX',
                'Orden': order,
                'RMSE': rmse,
                'MAE': mae,
                'MAPE': mape,
                'AIC': fit.aic,
                'BIC': fit.bic,
                'Ljung_Box_pvalue': lb['lb_pvalue'].values[0]
            })
        except:
            pass

    return resultados
```

```python
splits = {
    '80_20': 0.8,
    '90_10': 0.9,
    '95_5': 0.95
}

resultados_totales = {}

for nombre, split in splits.items():

    X_train, X_test, y_train, y_test = train_test_split_time_series(X, y, split)

    resultados = evaluar_modelos(X_train, X_test, y_train, y_test)

    df_res = pd.DataFrame(resultados).sort_values(by='RMSE')

    resultados_totales[nombre] = df_res

    print(f"\nResultados {nombre}")
    print(df_res.head(10))
```
