```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error

# ==========================================
# 1. PREPROCESAMIENTO Y CODIFICACIÓN DE FECHA
# ==========================================

# Asegurar que la columna 'fecha' sea reconocida como tipo Datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Ordenar el DataFrame cronológicamente (crucial para series temporales)
df = df.sort_values(by='fecha').reset_index(drop=True)

# Extraer variables numéricas a partir de la fecha
df['mes'] = df['fecha'].dt.month
df['dia_mes'] = df['fecha'].dt.day
df['dia_semana'] = df['fecha'].dt.dayofweek

# ==========================================
# 2. SELECCIÓN DE VARIABLES (X e y)
# ==========================================

# Tu variable objetivo
y = df['casos_ln']

# Variables predictoras (incluyendo las nuevas numéricas y EXCLUYENDO la 'fecha' original)
X = df[[
    'año', 'semana_epi', 'mes', 'dia_mes', 'dia_semana', 
    'prec_ln', 'dias_lluvia_ln', 'temp_bc', 'temp_max_bc', 
    'temp_min_bc', 'hum_esp_bc', 'hum_rel_bc', 'vel_vi_bc', 
    'vel_vi_max_bc', 'vel_vi_min_bc', 'uv_bc', 'soi_yj', 'sst_yj'
]]

# ==========================================
# 3. DIVISIÓN CRONOLÓGICA (Train / Test Split)
# ==========================================

# Definimos la proporción para el entrenamiento (80% entrenamiento, 20% prueba)
train_size = int(len(df) * 0.90)

# Dividimos los datos de manera secuencial (los primeros registros entrenan, los últimos evalúan)
X_train = X.iloc[:train_size]
X_test = X.iloc[train_size:]

y_train = y.iloc[:train_size]
y_test = y.iloc[train_size:]

# Opcional: Imprimir las fechas límite para verificar la división
fecha_corte = df['fecha'].iloc[train_size]
print(f"Datos de Entrenamiento: desde {df['fecha'].min().strftime('%Y-%m-%d')} hasta {fecha_corte.strftime('%Y-%m-%d')}")
print(f"Datos de Prueba (Test): desde {fecha_corte.strftime('%Y-%m-%d')} hasta {df['fecha'].max().strftime('%Y-%m-%d')}\n")

# ==========================================
# 4. ENTRENAMIENTO Y PREDICCIÓN
# ==========================================

# Configurar y entrenar el modelo
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train.ravel())

# Realizar predicciones
y_pred = model.predict(X_test)

# ==========================================
# 5. EVALUACIÓN DEL MODELO
# ==========================================
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)

print("--- Resultados de la Evaluación ---")
print(f"R² (Coeficiente de determinación): {r2:.4f}")
print(f"MAE (Error Absoluto Medio): {mae:.4f}")
print(f"MSE (Error Cuadrático Medio): {mse:.4f}")
print(f"MAPE (Error Porcentual Absoluto Medio): {mape:.4f}")
```

```python
import numpy as np
import pandas as pd

# ==========================================
# 1. DEFINIR LOS DATOS DEL CASO PUNTUAL
# ==========================================

# Supongamos que queremos evaluar una predicción para una fecha específica.
# Los datos deben ingresarse exactamente en el mismo orden y con los mismos nombres que 'X'.

datos_nuevos = {
    'año': [2024],
    'semana_epi': [22],
    'mes': [5],
    'dia_mes': [29],
    'dia_semana': [4],  # Viernes (0=Lunes, 6=Domingo)
    'prec_ln': [2.3],   # Ejemplo de valor transformado con logaritmo
    'dias_lluvia_ln': [1.6],
    'temp_bc': [26.5],  # Ejemplo de valor transformado con Box-Cox
    'temp_max_bc': [32.0],
    'temp_min_bc': [22.1],
    'hum_esp_bc': [18.2],
    'hum_rel_bc': [78.5],
    'vel_vi_bc': [4.2],
    'vel_vi_max_bc': [12.0],
    'vel_vi_min_bc': [1.1],
    'uv_bc': [8.5],
    'soi_yj': [-0.5],   # Ejemplo de valor transformado con Yeo-Johnson
    'sst_yj': [0.2]
}

# Convertir el diccionario en un DataFrame de una sola fila
X_puntual = pd.DataFrame(datos_nuevos)

# ==========================================
# 2. EJECUTAR LA PREDICCIÓN PUNTUAL
# ==========================================

# El modelo predice en la misma escala en la que fue entrenado (logarítmica: casos_ln)
prediccion_ln = model.predict(X_puntual)[0]

print("--- Evaluación de Predicción Puntual ---")
print(f"Predicción (en escala casos_ln): {prediccion_ln:.4f}")

# ==========================================
# 3. REGRESAR A LA ESCALA REAL (Opcional pero recomendado)
# ==========================================
# Como tu variable objetivo es 'casos_ln' (logaritmo natural de los casos), 
# para entender el número real de personas/casos debes aplicar la función exponencial.

# Nota: Si al transformar usaste log(casos + 1) para evitar el cero, usa np.expm1().
# Si usaste solo log(casos), usa np.exp(). Asumiremos np.expm1 por seguridad epidemiológica:
prediccion_real = np.expm1(prediccion_ln)

print(f"Predicción (en número real de casos): {round(prediccion_real)} casos")

# ==========================================
# 4. COMPARACIÓN SI TIENES EL DATO REAL
# ==========================================
# Si estás evaluando un registro que ya existía en tu dataset de prueba para verificar el error:

valor_real_ln = 1.95  # <--- Sustituye con el y_test real si lo tienes a la mano
valor_real_casos = np.expm1(valor_real_ln)

error_absoluto = abs(prediccion_real - valor_real_casos)

print(f"\nValor real (en escala casos_ln): {valor_real_ln:.4f}")
print(f"Valor real (en número real de casos): {round(valor_real_casos)} casos")
print(f"Diferencia absoluta: {error_absoluto:.2f} casos de error en esta predicción.")
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error, r2_score

# ==========================================
# 1. ESCALADO DE DATOS (Crucial para Redes Neuronales)
# ==========================================
# Las redes neuronales son muy sensibles a la escala. Random Forest no lo requería, pero aquí es obligatorio.
scaler_X = StandardScaler()

# Ajustamos el escalador SOLO con datos de entrenamiento para evitar fuga de información
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

# Convertir la variable objetivo a tipo float32 de numpy
y_train_nn = y_train.values.astype('float32')
y_test_nn = y_test.values.astype('float32')

# ==========================================
# 2. ARQUITECTURA DE LA RED NEURONAL
# ==========================================
# Diseñamos una red con Dropout y BatchNormalization para evitar el sobreajuste (overfitting)
model_nn = Sequential([
    # Capa de entrada y primera capa oculta
    Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    BatchNormalization(),
    Dropout(0.2),
    
    # Segunda capa oculta
    Dense(32, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    
    # Tercera capa oculta (para capturar interacciones complejas del clima)
    Dense(16, activation='relu'),
    
    # Capa de salida (1 sola neurona para regresión)
    Dense(1, activation='linear')
])

# Usamos el optimizador Adam y la pérdida Huber, que es más robusta a valores atípicos (outliers) que el MSE
model_nn.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.005),
    loss='huber', 
    metrics=['mae']
)

# ==========================================
# 3. ENTRENAMIENTO CON PARADA TEMPRANA (Early Stopping)
# ==========================================
# Si el modelo deja de mejorar en los datos de prueba, se detiene automáticamente
early_stop = EarlyStopping(
    monitor='val_loss', 
    patience=15, 
    restore_best_weights=True
)

history = model_nn.fit(
    X_train_scaled, y_train_nn,
    validation_data=(X_test_scaled, y_test_nn),
    epochs=150,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

# ==========================================
# 4. PREDICCIÓN Y EVALUACIÓN EN ESCALA REAL
# ==========================================
# 1. Predecir en la escala logarítmica
y_pred_nn_ln = model_nn.predict(X_test_scaled).flatten()

# 2. ¡CRUCIAL! Volver a la escala real (número de casos) antes de calcular el MAPE.
# Si tu variable era casos_ln = ln(casos + 1), usamos expm1.
y_test_real = np.expm1(y_test_nn)
y_pred_real = np.expm1(y_pred_nn_ln)

# Reemplazar cualquier valor negativo accidental en la predicción real por 0
y_pred_real = np.clip(y_pred_real, 0, None)

# 3. Calcular métricas reales
# Para evitar que las semanas con 0 casos reales lleven el MAPE al infinito,
# calculamos el MAPE filtrando o usando un pequeño ajuste, o evaluamos con MAE.
mae_real = mean_absolute_error(y_test_real, y_pred_real)
r2_real = r2_score(y_test_real, y_pred_real)

# MAPE calculado de forma estándar sobre los casos reales
mape_real = mean_absolute_percentage_error(y_test_real, y_pred_real)

print("\n--- Resultados de la Red Neuronal (En Casos Reales) ---")
print(f"R² Score: {r2_real:.4f}")
print(f"MAE (Error promedio en número de personas): {mae_real:.2f} casos")
print(f"MAPE Real: {mape_real * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# ==========================================
# SCRIPT PARA GRAFICAR EL DESEMPEÑO DE LA RED
# ==========================================

# Crear la figura con dos subtramas (Loss y MAE)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# 1. Gráfico de la Función de Pérdida (Huber Loss)
ax1.plot(history.history['loss'], label='Entrenamiento (Train Loss)', color='#1f77b4', linewidth=2)
ax1.plot(history.history['val_loss'], label='Testeo (Validation Loss)', color='#ff7f0e', linewidth=2)
ax1.set_title('Evolución de la Función de Pérdida (Huber Loss)', fontsize=12, fontweight='bold')
ax1.set_xlabel('Épocas', fontsize=10)
ax1.set_ylabel('Pérdida', fontsize=10)
ax1.legend(fontsize=10)
ax1.grid(True, linestyle='--', alpha=0.6)

# 2. Gráfico del Error Absoluto Medio (MAE)
ax2.plot(history.history['mae'], label='Entrenamiento (Train MAE)', color='#2ca02c', linewidth=2)
ax2.plot(history.history['val_mae'], label='Testeo (Validation MAE)', color='#d62728', linewidth=2)
ax2.set_title('Evolución del Error Absoluto Medio (MAE)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Épocas', fontsize=10)
ax2.set_ylabel('MAE (Escala Logarítmica)', fontsize=10)
ax2.legend(fontsize=10)
ax2.grid(True, linestyle='--', alpha=0.6)

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error

# ==========================================================
# 1. PREPROCESAMIENTO, CODIFICACIÓN Y DIVISIÓN CRONOLÓGICA
# ==========================================================

# Asegurar formato de fecha y ordenar cronológicamente
df['fecha'] = pd.to_datetime(df['fecha'])
df = df.sort_values(by='fecha').reset_index(drop=True)

# Extracción de características de la fecha
df['mes'] = df['fecha'].dt.month
df['dia_mes'] = df['fecha'].dt.day
df['dia_semana'] = df['fecha'].dt.dayofweek

# Selección de variables
y = df['casos_ln']
X = df[[
    'año', 'semana_epi', 'mes', 'dia_mes', 'dia_semana', 
    'prec_ln', 'dias_lluvia_ln', 'temp_bc', 'temp_max_bc', 
    'temp_min_bc', 'hum_esp_bc', 'hum_rel_bc', 'vel_vi_bc', 
    'vel_vi_max_bc', 'vel_vi_min_bc', 'uv_bc', 'soi_yj', 'sst_yj'
]]

# División cronológica (80% entrenamiento, 20% testeo)
train_size = int(len(df) * 0.80)
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

# Escalado de datos (Obligatorio para redes neuronales)
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

y_train_nn = y_train.values.astype('float32')
y_test_nn = y_test.values.astype('float32')

# ==========================================================
# 2. ARQUITECTURA REDISEÑADA (MÁS SIMPLE Y REGULARIZADA)
# ==========================================================

model_nn = Sequential([
    # Capa de entrada + 1ra Capa Oculta reducida a 32 neuronas con regularización L2
    Dense(32, activation='relu', kernel_regularizer=l2(0.01), input_shape=(X_train_scaled.shape[1],)),
    BatchNormalization(),
    Dropout(0.5),  # Incrementado a 50% para mitigar memorización
    
    # 2da Capa Oculta reducida a 16 neuronas con regularización L2
    Dense(16, activation='relu', kernel_regularizer=l2(0.01)),
    BatchNormalization(),
    Dropout(0.5),  # Incrementado a 50%
    
    # Capa de salida lineal
    Dense(1, activation='linear')
])

# Compilación manteniendo la pérdida Huber por robustez
model_nn.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.005),
    loss='huber', 
    metrics=['mae']
)

# Early Stopping para rescatar la mejor época
early_stop = EarlyStopping(
    monitor='val_loss', 
    patience=15, 
    restore_best_weights=True
)

# ==========================================================
# 3. ENTRENAMIENTO
# ==========================================
history = model_nn.fit(
    X_train_scaled, y_train_nn,
    validation_data=(X_test_scaled, y_test_nn),
    epochs=150,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

# ==========================================================
# 4. EVALUACIÓN EXCLUSIVA DE LA MÉTRICA MAE
# ==========================================================

# Predicciones en escala logarítmica
y_pred_nn_ln = model_nn.predict(X_test_scaled).flatten()

# Calcular MAE en la escala en que entrenó la red (Logarítmica)
mae_logaritmico = mean_absolute_error(y_test_nn, y_pred_nn_ln)

# Transformación inversa a la escala real (Casos de Dengue)
y_test_real = np.expm1(y_test_nn)
y_pred_real = np.expm1(y_pred_nn_ln)
y_pred_real = np.clip(y_pred_real, 0, None)  # Evitar casos negativos teóricos

# Calcular MAE en la escala real de personas
mae_real = mean_absolute_error(y_test_real, y_pred_real)

print("\n=======================================================")
print("          MÉTRICAS DE DESEMPEÑO SOLICITADAS            ")
print("=======================================================")
print(f"MAE (En escala transformada 'casos_ln'): {mae_logaritmico:.4f}")
print(f"MAE REAL (Error promedio en número de personas): {mae_real:.2f} casos de dengue")
print("=======================================================\n")

# ==========================================================
# 5. GRÁFICOS COMPARATIVOS DE DESEMPEÑO
# ==========================================================

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Panel 1: Comparativa de la Función de Pérdida
ax1.plot(history.history['loss'], label='Entrenamiento (Train Huber Loss)', color='#1f77b4', linewidth=2)
ax1.plot(history.history['val_loss'], label='Testeo (Validation Huber Loss)', color='#ff7f0e', linewidth=2)
ax1.set_title('Comparación de Pérdida (Huber Loss)', fontsize=12, fontweight='bold')
ax1.set_xlabel('Épocas', fontsize=10)
ax1.set_ylabel('Pérdida', fontsize=10)
ax1.legend(fontsize=10)
ax1.grid(True, linestyle='--', alpha=0.5)

# Panel 2: Comparativa de la Métrica MAE
ax2.plot(history.history['mae'], label='Entrenamiento (Train MAE)', color='#2ca02c', linewidth=2)
ax2.plot(history.history['val_mae'], label='Testeo (Validation MAE)', color='#d62728', linewidth=2)
ax2.set_title('Comparación del Error Absoluto Medio (MAE)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Épocas', fontsize=10)
ax2.set_ylabel('MAE (Escala Logarítmica)', fontsize=10)
ax2.legend(fontsize=10)
ax2.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

¡Excelente! Estas nuevas gráficas muestran un cambio radical y sumamente positivo. Las técnicas de regularización (el aumento del Dropout, la reducción de neuronas y la penalización L2) funcionaron exactamente como se esperaba teóricamente.

Aquí tienes la interpretación técnica y detallada de lo que reflejan tus nuevos gráficos de rendimiento:

## 1. El Diagnóstico Principal: Aprendizaje Exitoso y Estable

A diferencia de las primeras gráficas donde las curvas se separaban con violencia a partir de la época 3, aquí observamos un comportamiento clásico de **convergencia saludable**:

* **Alineación de curvas:** Las líneas de entrenamiento (azul/verde) y las de testeo (naranja/roja) descienden **juntas** y con una tendencia paralela a lo largo del tiempo.
* **Control del sobreajuste:** Aunque persiste una pequeña brecha natural entre el entrenamiento y el testeo (lo cual es normal porque el modelo siempre tenderá a rendir ligeramente mejor con los datos que ya conoce), el sobreajuste catastrófico que destruía tus predicciones ha desaparecido.

## 2. Un Proceso de Aprendizaje más Lento pero Seguro

* Ahora, el modelo se tomó el tiempo de explorar y aprender de manera gradual durante más de **60 épocas**.
* Al apagar aleatoriamente el 50% de las neuronas en cada paso (Dropout), obligaste a la red a no depender de "atajos" ni a memorizar datos específicos. * El mecanismo de **Early Stopping** actuó de manera perfecta aquí: detectó que después de esos mínimos el modelo ya no estaba mejorando genuinamente en el conjunto de prueba, detuvo el entrenamiento en la época 64 y restauró los pesos matemáticos de esa época dorada (alrededor de la 49).
