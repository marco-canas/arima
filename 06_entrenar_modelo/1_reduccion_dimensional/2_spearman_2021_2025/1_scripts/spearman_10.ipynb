Como los valores de correlación son bajos, pon los umbrales de correlación de acuerdo a los dos mejores percentiles. Pero para la reducción dimensional, solo debes considerar a las variables meteorológicas, excluye al `año`, `semana_epi` y `fecha` para la reducción, pero incluyelas en el dataset final, junto a la variable objetivo de `casos_dengue`. el dataset a reducir se encuentra en la ubicación "C:\Users\marco\Documentos\investigacion\arima\06_entrenar_modelo\1_reduccion_dimensional\2_spearman_2021_2025\2_data\1_raw\1_meteo_epi_rezagos.xlsx", y debes guardar el dataset reducido, como un archivo .xlsx en las ubicaciones: "C:\Users\marco\Documentos\investigacion\arima\06_entrenar_modelo\3_arimax\1_criterio_aicc\3_rezagos_spiarman_10_aicc\2_datos\1_raw" y en "C:\Users\marco\Documentos\investigacion\arima\06_entrenar_modelo\1_reduccion_dimensional\2_spearman_2021_2025\2_data\2_processed"


El script carga los datos, calcula las correlaciones con respecto a `casos_dengue`, filtra las variables meteorológicas según los percentiles más altos de correlación con la variable objetivo, y luego reduce la redundancia (multicolinealidad) eliminando variables altamente correlacionadas entre sí, seleccionando exactamente las 10 mejores variables meteorológicas. Finalmente, guarda el conjunto de datos resultante en las dos rutas especificadas manteniendo las columnas de control (`fecha`, `año`, `semana_epi` y `casos_dengue`).

```python
import os
import pandas as pd
import numpy as np

# 1. Definición de rutas de entrada y salida
# Cambia esta ruta a la ubicación real de tu archivo de entrada si es necesario
ruta_entrada = r"C:\Users\marco\Documentos\investigacion\arima\06_entrenar_modelo\1_reduccion_dimensional\2_spearman_2021_2025\2_data\1_raw\1_meteo_epi_rezagos.xlsx" 

ruta_salida_1 = r"C:\Users\marco\Documentos\investigacion\arima\06_entrenar_modelo\1_reduccion_dimensional\2_spearman_2021_2025\2_data\2_processed"
ruta_salida_2 = r"C:\Users\marco\Documentos\investigacion\arima\06_entrenar_modelo\3_arimax\1_criterio_aicc\3_rezagos_spiarman_10_aicc\2_datos\1_raw"

nombre_archivo_salida = "1_spearman_10.xlsx"

# Asegurar que los directorios de salida existan
os.makedirs(ruta_salida_1, exist_ok=True)
os.makedirs(ruta_salida_2, exist_ok=True)

# 2. Cargar los datos
print("Cargando datos...")
df = pd.read_excel(ruta_entrada)

# Definir variables de control y objetivo
variables_excluidas = ['fecha', 'año', 'semana_epi']
variable_objetivo = 'casos_dengue'

# Identificar variables meteorológicas (con y sin rezago)
variables_meteorologicas = [col for col in df.columns if col not in variables_excluidas + [variable_objetivo]]

# 3. Calcular la matriz de correlación de Spearman
print("Calculando correlaciones de Spearman...")
matriz_corr = df[[variable_objetivo] + variables_meteorologicas].corr(method='spearman')

# Obtener la correlación absoluta con la variable objetivo
corr_con_objetivo = matriz_corr[variable_objetivo].abs().drop(variable_objetivo)

# 4. Filtrado inicial por percentil (Atributos mejor correlacionados con la variable objetivo)
# Como los valores son bajos, tomamos el percentil 80 (el top 20% de las variables con mayor correlación)
umbral_percentil = corr_con_objetivo.quantile(0.80)
variables_candidatas = corr_con_objetivo[corr_con_objetivo >= umbral_percentil].index.tolist()

print(f"Número de variables candidatas tras aplicar percentil (umbral r > {umbral_percentil:.4f}): {len(variables_candidatas)}")

# 5. Selección de 10 variables minimizando la correlación entre sí (Evitar multicolinealidad)
# Ordenamos las candidatas de mayor a menor correlación con el dengue
variables_ordenadas = corr_con_objetivo[variables_candidatas].sort_values(ascending=False).index.tolist()

variables_seleccionadas = []

# Iteramos sobre las variables ordenadas por relevancia y filtramos las que están muy correlacionadas entre sí
for var in variables_ordenadas:
    if len(variables_seleccionadas) == 10:
        break
    
    # Comprobar la correlación de 'var' con las ya seleccionadas
    if len(variables_seleccionadas) == 0:
        variables_seleccionadas.append(var)
    else:
        # Evaluamos la correlación máxima con las ya elegidas
        corr_inter_var = df[variables_seleccionadas + [var]].corr(method='spearman').loc[var, variables_seleccionadas].abs()
        
        # Ponemos un umbral dinámico basado en percentiles inter-variable si es necesario, 
        # o simplemente nos quedamos con las que tengan menor correlación entre sí.
        # Para asegurar que obtenemos exactamente 10 de manera robusta:
        umbral_inter_var = 0.75  # Evita meter variables que compartan más del 75% de correlación (ajustable)
        
        if corr_inter_var.max() < umbral_inter_var:
            variables_seleccionadas.append(var)

# Si por el umbral estricto no llegamos a 10, completamos con las siguientes mejores con menor correlación posible
if len(variables_seleccionadas) < 10:
    for var in variables_ordenadas:
        if var not in variables_seleccionadas:
            variables_seleccionadas.append(var)
        if len(variables_seleccionadas) == 10:
            break

print("\nLas 10 variables meteorológicas seleccionadas son:")
for i, v in enumerate(variables_seleccionadas, 1):
    print(f"{i}. {v} (Corr con dengue: {corr_con_objetivo[v]:.4f})")

# 6. Construir el dataset final
columnas_finales = variables_excluidas + variables_seleccionadas + [variable_objetivo]
df_final = df[columnas_finales]

# 7. Guardar el archivo en las dos ubicaciones especificadas
camino_completo_1 = os.path.join(ruta_salida_1, nombre_archivo_salida)
camino_completo_2 = os.path.join(ruta_salida_2, nombre_archivo_salida)

df_final.to_excel(camino_completo_1, index=False)
df_final.to_excel(camino_completo_2, index=False)

print(f"\nProceso completado con éxito.")
print(f"Dataset guardado en Ubicación 1: {camino_completo_1}")
print(f"Dataset guardado en Ubicación 2: {camino_completo_2}")
print(f"Dimensiones del dataset guardado: {df_final.shape}")
```


# Explicación de los puntos clave del script:

1. **Exclusión e Inclusión:** Separa desde el inicio las variables de control (`fecha`, `año`, `semana_epi`) y la variable objetivo (`casos_dengue`) de la matriz de meteorología para que no interfieran en el cálculo de reducción dimensional, pero las reincorpora intactas al generar el archivo final.
2. **Criterio de Percentiles:** Para lidiar con los valores bajos de correlación general, utiliza `.quantile(0.80)` sobre las correlaciones absolutas de Spearman. Esto significa que solo toma el "Top 20%" superior de las variables con mejor relación con el dengue.
3. **Control de Multicolinealidad (menos correlacionadas entre sí):** El script ordena las variables candidatas de mayor a menor fuerza explicativa y va agregándolas una a una, asegurándose de que la correlación interna con las variables meteorológicas ya seleccionadas no supere un umbral (establecido de forma estándar en 0.75). De esta manera se remueve la redundancia informativa de los rezagos continuos (por ejemplo, si `temp_lag_1` y `temp_lag_2` son casi idénticas, solo seleccionará una).
4. **Exportación automatizada:** Utiliza `os.makedirs(..., exist_ok=True)` para garantizar que si las carpetas profundas de tus rutas de investigación no existen, Python las creará automáticamente antes de intentar guardar el archivo `.csv`.


Dado que las correlaciones generales son bajas, el script calcula dinámicamente los **percentiles más altos** para tomar decisiones automatizadas:

1. **Filtro de Relevancia:** Selecciona las variables meteorológicas que tengan una correlación con `casos_dengue` por encima del **percentil 90** (el top 10% de las mejores correlaciones).
2. **Filtro de Redundancia:** Entre las variables seleccionadas, evalúa su correlación interna (multicolinealidad). Si dos variables están muy correlacionadas entre sí (superando el **percentil 25** de las correlaciones del set filtrado), elimina la que tenga menor impacto sobre los casos de dengue.

# Script de Python


```python
import pandas as pd
import numpy as np

# 1. Cargar el dataset (ajusta la ruta según corresponda)
# Nota: Como especificaste archivo .xlsx, usamos read_excel. Si usas el .csv, cambia a read_csv.
df = pd.read_excel(r"C:\Users\marco\Documentos\investigacion\arima\07_seleccion_entrenar_modelo\1_reduccion_dimensional\2_spearman_2021_2025\2_data\1_raw\1_meteo_epi_rezagos.xlsx")

# 2. Definir variables
target = "casos_dengue"

# Excluir variables de control/tiempo para quedarnos solo con las meteorológicas (y sus lags)
exclude_cols = ["fecha", "año", "semana_epi", "soi", "sst", target]
features = [col for col in df.columns if col not in exclude_cols]

# Asegurar que los datos sean numéricos y no tengan nulos para el cálculo de correlación
df_numeric = df[features + [target]].dropna()

# 3. Calcular la matriz de correlación completa
corr_matrix = df_numeric.corr(method="pearson")

# --- PASO 1: Filtrar por relevancia con la variable objetivo ---
# Obtener la correlación absoluta con 'casos_dengue'
target_corr = corr_matrix[target].abs().drop(target)

# Definir el umbral usando el percentil alto (ej. Percentil 90 = Top 10% mejores predictores)
threshold_target = target_corr.quantile(0.60)
selected_features = target_corr[target_corr >= threshold_target].index.tolist()

print(f"--- PASO 1: SELECCIÓN POR RELEVANCIA ---")
print(f"Umbral de correlación con el objetivo (Percentil 90): {threshold_target:.4f}")
print(f"Atributos seleccionados ({len(selected_features)}): {selected_features}\n")

# --- PASO 2: Filtrar por redundancia (menos correlacionados entre sí) ---
# Matriz de correlación solo entre las variables preseleccionadas
sub_corr_matrix = df_numeric[selected_features].corr().abs()

# Extraer los valores de la diagonal superior (para no duplicar parejas ni evaluar la diagonal de 1s)
upper_tri = sub_corr_matrix.where(np.triu(np.ones(sub_corr_matrix.shape), k=1).astype(bool))
all_corrs = upper_tri.stack().values

# Definir el umbral de redundancia usando un percentil bajo de las correlaciones existentes
# Como las correlaciones son bajas, usamos el percentil 25 de sus intercorrelaciones para ser estrictos
threshold_redundancy = np.nanpercentile(all_corrs, 90)

print(f"--- PASO 2: FILTRADO POR REDUNDANCIA ---")
print(f"Umbral máximo permitido entre atributos (Percentil 25): {threshold_redundancy:.4f}")

# Algoritmo para eliminar características redundantes
features_to_remove = set()
for col in sub_corr_matrix.columns:
    if col in features_to_remove:
        continue
    # Buscar variables altamente correlacionadas con la variable actual
    redundant_cols = sub_corr_matrix.index[(sub_corr_matrix[col] > threshold_redundancy) & (sub_corr_matrix.index != col)].tolist()
    
    for r_col in redundant_cols:
        if r_col not in features_to_remove:
            # Comparar cuál tiene peor correlación con el dengue y eliminarla
            if target_corr[col] >= target_corr[r_col]:
                features_to_remove.add(r_col)
            else:
                features_to_remove.add(col)
                break  # Si la variable actual es eliminada, dejamos de evaluarla

final_features = [f for f in selected_features if f not in features_to_remove]

print(f"Atributos eliminados por alta redundancia: {list(features_to_remove)}")
print(f"\n--- RESULTADO FINAL ---")
print(f"Atributos óptimos finales ({len(final_features)}): {final_features}")

# 4. Crear el nuevo dataset con la reducción dimensional realizada
df_reducido = df[["fecha", "año", "semana_epi", target] + final_features]
df_reducido.to_csv(r"C:\Users\marco\Documentos\investigacion\arima\07_seleccion_entrenar_modelo\1_reduccion_dimensional\2_spearman_2021_2025\2_data\2_processed\2_meteo_epi_rezagos_spearman.csv", index=False)
```



# ¿Cómo funciona la lógica de percentiles instalada?

* **`target_corr.quantile(0.90)`**: Encuentra el valor de correlación que supera al **90%** de las demás variables con respecto al dengue. Esto garantiza que solo te quedas con la "crema y nata" de tus atributos climáticos (con o sin rezago).
* **`np.percentile(all_corrs, 25)`**: Al analizar cómo se correlacionan estas variables seleccionadas entre sí, calcula el percentil 25. Al usar este valor como límite máximo permitido, el script se vuelve muy estricto: si dos variables se correlacionan un poco más de lo normal (rebasando ese primer cuartil), el script descarta inmediatamente la que tenga menor impacto sobre los `casos_dengue`.
